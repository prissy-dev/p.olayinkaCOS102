```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
print(df)
```

```python
df.info()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
    df.head()


```

```python

   print(df.columns)
```

```python
print(df.isnull().sum())

```

```python
df.to_csv("Cleaned titanic.csv", index=False)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['Age'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure()
df.groupby('Sex')['Survived'].sum().plot(kind='bar')
plt.title("Survival Count by Gender")
plt.xlabel("Sex")
plt.ylabel("Number Survived")
plt.show()
```

```python
 plt.figure()
df['Embarked'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Passengers by Embarkation Port")
plt.ylabel("")
plt.show()
```

For the first visualization, most of the passengers are between zero and thirty years, and a few are above the range.
For the second visualization, females survived the disaster more than the males.
For the third visualization, more passengers embarked on port S,folowed by port C,port Q has the least percentage of passengers embarked.

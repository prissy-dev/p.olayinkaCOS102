```python
import tkinter as tk
from tkinter import messagebox, ttk
import pandas as pd

# LOGIC SECTION 

def get_departments():
    try:
        df = pd.read_csv('GIG-logistics.csv')
        
        return sorted(df['Department'].unique().tolist())
    except Exception:
        return ["Logistics", "Admin", "Accounts", "Dispatch"] 

def verify_employee():
    name_input = name_entry.get().strip()
    dept_input = dept_dropdown.get() 

    if not name_input or not dept_input:
        messagebox.showwarning("Input Error", "Please enter your name and select a department.")
        return

    else:
        df = pd.read_csv('GIG-logistics.csv')
        
        # Check for match
        match = df[(df['Name'].str.lower() == name_input.lower()) & 
                   (df['Department'] == dept_input)]

        if not match.empty:
            # Find colleagues in the same department
            colleagues = df[df['Department'] == dept_input]['Name'].tolist()
            if name_input.title() in colleagues:
                colleagues.remove(name_input.title())
            
            show_welcome_window(name_input.title(), dept_input, colleagues)
        else:
            messagebox.showerror("Access Denied", f"No record of '{name_input}' in {dept_input}.")
            
    except FileNotFoundError:
      messagebox.showerror("System Error", "Database file (GIG-logistics.csv) not found!")

def show_welcome_window(name, dept, colleagues):
    welcome_win = tk.Toplevel(root)
    welcome_win.title("GIG Employee Portal")
    welcome_win.geometry("400x450")
    welcome_win.configure(bg="#ffffff")

    # Success UI
    tk.Label(welcome_win, text="✓ Verification Successful", fg="green", bg="white").pack(pady=10)
    tk.Label(welcome_win, text=f"Welcome, {name}"), bg="white".pack()
    tk.Label(welcome_win, text=f"Department: {dept}"), bg="white", fg="gray".pack(pady=5)
    
    tk.Label(welcome_win, text="Other Members in your Department: "), bg="white".pack(pady=(20, 5))
    
    # Listbox with a scrollbar
    frame = tk.Frame(welcome_win)
    frame.pack(pady=5)
    
    scrollbar = tk.Scrollbar(frame)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
    
    listbox = tk.Listbox(frame, width=40, height=10, yscrollcommand=scrollbar.set)
    for person in colleagues:
        listbox.insert(tk.END, f"  • {person}")
    listbox.pack()
    
    scrollbar.config(command=listbox.yview)
    ttk.Button(welcome_win, text="Logout", command=welcome_win.destroy).pack(pady=20)



root = tk.Tk()
root.title("GIG Logistics Identity System")
root.geometry("400x450")
root.eval('tk::PlaceWindow . center') 

# Header Logo Style
tk.Label(root, text="GIG LOGISTICS"), fg="#e67e22".pack(pady=(30, 0))
tk.Label(root, text="EMPLOYEE VERIFICATION"), fg="gray".pack(pady=(0, 20))

# Name Input
tk.Label(root, text="Enter Full Name:").pack( padx=50)
name_entry = ttk.Entry(root, width=35)
name_entry.pack(pady=5, padx=50, ipady=3)

# Department Dropdown
tk.Label(root, text="Select Department:").pack(padx=50, pady=(10, 0))
dept_dropdown = ttk.Combobox(root, width=32)
dept_dropdown['values'] = get_departments()
dept_dropdown.pack(pady=5, padx=50, ipady=3)

# Buttons
btn_frame = tk.Frame(root)
btn_frame.pack(pady=30)

verify_btn = tk.Button(btn_frame, text="VERIFY", bg="#2ecc71", fg="white", 
                       width=15, command=verify_employee)
verify_btn.grid(row=0, column=0, padx=5)

clear_btn = tk.Button(btn_frame, text="CLEAR", bg="#95a5a6", fg="white", 
                         width=10, command=lambda: name_entry.delete(0, tk.END))
clear_btn.grid(row=0, column=1, padx=5)

root.mainloop()
```

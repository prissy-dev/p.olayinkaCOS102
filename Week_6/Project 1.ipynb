```python
import tkinter as tk
import random as rd

class Employee:
    def __init__(self):
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide",
            "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]

        self.tasks = [
            "Loading", "Transporting", "Reviewing Orders",
            "Customer Service", "Delivering Items"
        ]

    def check_employee(self, name):
        return name in self.employees

    def take_attendance(self, name):
        return f"Attendance taken for {name}"

    def assign_task(self):
        return rd.choice(self.tasks)

    def refuse_access(self):
        return "Sorry, we are unable to verify your employee details at the moment."

def submit():
    name = entry.get()
    
    emp = Employee()

    if emp.check_employee(name):
        attendance = emp.take_attendance(name)
        task = emp.assign_task()
        result_label.config(
            text=f"{attendance}\nTask Assigned: {task}",
            fg="green"
        )
    else:
        result_label.config(
            text=emp.refuse_access(),
            fg="red"
        )

root = tk.Tk()
root.title("Employee System")
root.geometry("400x300")

tk.Label(root, text="Enter Your Name").pack(pady=10)

entry = tk.Entry(root)
entry.pack(pady=5)

tk.Button(root, text="Submit", command=submit).pack(pady=10)

result_label = tk.Label(root, text="")
result_label.pack(pady=20)

root.mainloop()
```

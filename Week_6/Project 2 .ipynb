```python
import tkinter as tk

class Delivery:
    def calculate_price(self, location, weight):
        if location.lower() == "pau":
            if weight >= 10:
                return 2000
            else:
                return 1500

        elif location.lower() == "epe":
            if weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None


def calculate():
    location = location_entry.get()
    weight = float(weight_entry.get())

    delivery = Delivery()
    price = delivery.calculate_price(location, weight)

    if price:
        result_label.config(
            text=f"Delivery Cost: ₦{price}",
            fg="green"
        )
    else:
        result_label.config(
            text="Invalid Location!",
            fg="red"
        )
        
root = tk.Tk()
root.title("Delivery Calculator")
root.geometry("400x300")

tk.Label(root, text="Enter Location (PAU or Epe)").pack(pady=5)
location_entry = tk.Entry(root)
location_entry.pack()

tk.Label(root, text="Enter Weight (kg)").pack(pady=5)
weight_entry = tk.Entry(root)
weight_entry.pack()

tk.Button(root, text="Calculate", command=calculate).pack(pady=10)

result_label = tk.Label(root, text="")
result_label.pack(pady=20)

root.mainloop()
```

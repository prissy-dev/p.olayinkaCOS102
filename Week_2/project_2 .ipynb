```python
# Importing Libraries
import pandas as pd
import numpy as np

# List the equations
print("Select the type of equation")
print("1:Quadratic")
print("2:Cubic")
print("3:Quartic")

# Asking users to choose
select = int(input("Enter your option: "))

if select == 1:
    a = float(input("Enter value of A: "))
    b = float(input("Enter value of B: "))
    c = float(input("Enter value of C: "))

    d = b*b - 4*a*c

    w1 = (-b + d**0.5) / (2*a)
    w2 = (-b - d**0.5) / (2*a)

    roots = np.roots([a,b,c])
    print("The roots are:", w1, w2)

elif select == 2:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))

    roots = np.roots([a,b,c,d])
    print("The roots of the cubic equation are:", roots)

elif select == 3:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    roots = np.roots([a,b,c,d,e])
    print("The roots of the quartic equation are:", roots)

else:
    print("Invalid selection,try again next time!")
```

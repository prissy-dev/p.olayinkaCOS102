```python
import tkinter as tk
from tkinter import messagebox

class Person:
    def __init__(self, name):
        self.name = name
        

    def display(self):
        return f"Name: {self.name}"

class Bank:
    def common_services(self):
        return [
            "Loans",
            "Investment accounts",
            "Insurance services"
        ]



class Employee(Person, Bank):
    def __init__(self, name, division):
        super().__init__(name)
        self.division = division


class Retail(Employee):
    def services(self):
        return self.common_services() + [
            "Savings account",
            "Current account",
            "Personal loans"
        ]


class Global(Employee):
    def services(self):
        return self.common_services() + [
            "Foreign accounts",
            "Currency exchange",
            "International transfers"
        ]


class Commercial(Employee):
    def services(self):
        return self.common_services() + [
            "Business advisory"
        ]


def show_result():
    name = name_input.get()
    division = division_input.get().lower()

    if division == "retail":
        emp = Retail(name, division)
    elif division == "global":
        emp = Global(name, division)
    elif division == "commercial":
        emp = Commercial(name, division)
    else:
        messagebox.showerror("Error", "Please enter a valid division")
        return

    result = emp.display() + f"\nDivision: {division.capitalize()}\nServices:"

    for item in emp.services():
        result += "- " + item + "\n"

    messagebox.showinfo("Details", result)

root = tk.Tk()
root.title("Bank Service System")
root.geometry("400x300")

tk.Label(root, text="Enter Name").pack()
name_input = tk.Entry(root)
name_input.pack()

tk.Label(root, text="Enter Division (Retail/Global/Commercial)").pack()
division_input = tk.Entry(root)
division_input.pack()

tk.Button(root, text="Show Details", command=show_result).pack(pady=10)

root.mainloop()
```

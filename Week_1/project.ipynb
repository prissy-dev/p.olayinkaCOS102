```python
# Calculation for simple interest
P = float(input("Let the principal be:"))
R = float(input("Let the rate in decimal be:"))
T = float(input("Let the time in years be:"))
A = P * (1 + (R/100) * T)
print("The amount is:", A)
                
                
```

```python
# Calculation for compound interest
P = float(input("Let the principal be:"))
R = float(input("Let the rate in decimal be:"))
t = float(input("Let the time in years be:"))
n = int(input("Let the number of times interest per year:"))
A = P * (1 + R/n) ** (n*t)
print("The amount is:", A)
                
```

```python
# Calculation for Annuity plan
PMT = float(input("Let the payment amount be:"))
R = float(input("Let the rate in decimal be:"))
t = float(input("Let the time in years be:"))
n = int(input("Let the number of times compounded per year:"))
A = PMT * (((1 + R/n) ** (n*t) - 1) / (R/n))
print("The amount is:", A)
```

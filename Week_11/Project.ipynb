```python
import pandas as pd
from IPython.display import display

class PAUStudentClassifier:
    def __init__(self, file_path):
        
        self.df = pd.read_csv(file_path)
        
        self.__security_exceptions = []

    def process_records(self):
        
        pirates = self.df[(self.df['Age'] > 14) & (self.df['Age'] < 18)]
        pirates.to_csv('The_Pirates.csv', index=False)
        print("THE PIRATES")
        print(pirates if not pirates.empty else "No students found in this category.", "\n")

        
        yankees = self.df[(self.df['Age'] > 18) & (self.df['Age'] < 22)]
        yankees.to_csv('The_Yankees.csv', index=False)
        print("THE YANKEES")
        print(yankees if not yankees.empty else "No students found in this category.", "\n")

        
        bulls = self.df[(self.df['Age'] > 22) & (self.df['Age'] < 25)]
        bulls.to_csv('The_Bulls.csv', index=False)
        print("THE BULLS")
        print(bulls if not bulls.empty else "No students found in this category.", "\n")

        
        over_age = self.df[self.df['Age'] > 25]
        if not over_age.empty:
            for index, row in over_age.iterrows():
                error_msg = f"Security Exception: Access Restricted for {row['Name']} (Age: {row['Age']})."
                self.__security_exceptions.append(error_msg)

    
    def get_security_exceptions(self):
        if self.__security_exceptions:
            return "\n".join(self.__security_exceptions)
        return "No security exceptions detected."




if __name__ == "__main__":
    
    classifier = PAUStudentClassifier('SIS.csv')
    classifier.process_records()
    
    print("ENCAPSULATION REPORT")
    print(classifier.get_security_exceptions())
```

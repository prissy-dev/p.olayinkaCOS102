```python
from abc import ABC, abstractmethod


class coup_de_ecriva(ABC):
    
    @abstractmethod
    def Fan_Page(self):
        pass


class FC_Cirok(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to FC Cirok")

class Madiba_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Madiba FC")

class Blue_Jay_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Blue Jay FC")

class TSG_Walker(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to TSG Walker")


def main():
    print("--- Club Fan Page Portal ---")
    user_input = input("Enter the name of the club you support: ").strip()
    
    
    club_mapping = {
        "FC Cirok": FC_Cirok,
        "Madiba FC": Madiba_FC,
        "Blue Jay FC": Blue_Jay_FC,
        "TSG Walker": TSG_Walker
    }
    
    
    if user_input in club_mapping:
        club_class = club_mapping[user_input]
        club_object = club_class()  
        club_object.Fan_Page()      
    else:
        print("Error: Club not found. Please type the name exactly (e.g., 'FC Cirok').")

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox, ttk


class Vendor:
    def __init__(self, name, menu_dict):
        self.name = name
        self.menu_dict = menu_dict  # Dictionary format: {"Item": Price}

class VendorManager:
    def __init__(self):
        # Initialize the vendor database exactly as shown in the image tables
        self.vendors = {
            "Cooperative Cafeteria": Vendor("Cooperative Cafeteria", {
                "Jollof Rice and Stew": 200,
                "White Rice and Stew": 200,
                "Fried Rice": 200,
                "Salad": 100,
                "Plantain": 100
            }),
            "Faith Hostel Cafeteria": Vendor("Faith Hostel Cafeteria", {
                "Fried Rice": 400,
                "White Rice and Stew": 400,
                "Jollof Rice": 400,
                "Beans": 200,
                "Chicken": 1000
            }),
            "Student Centre Cafeteria": Vendor("Student Centre Cafeteria", {
                "Chicken Fried Rice": 800,
                "Pomo Sauce": 300,
                "Spaghetti Jollof": 500,
                "Amala/Ewedu": 500,
                "Semo with Eforiro Soup": 500
            })
        }
    
    def get_vendor_names(self):
        return list(self.vendors.keys())
    
    def get_vendor_menu(self, name):
        if name in self.vendors:
            return self.vendors[name].menu_dict
        return None


class VendorApp:
    def __init__(self, root):
        self.manager = VendorManager()
        
        self.root = root
        self.root.title("PAU External Food Vendors Management")
        self.root.geometry("500x450")
        self.root.config(bg="#f4f4f4")
        
        
        self.title_label = tk.Label(
            root, text="PAU External Food Vendors", 
            font=("Helvetica", 16, "bold"), bg="#f4f4f4", fg="#333"
        )
        self.title_label.pack(pady=15)
        
        
        self.selection_frame = tk.Frame(root, bg="#f4f4f4")
        self.selection_frame.pack(pady=10)
        
        self.select_label = tk.Label(
            self.selection_frame, text="Select Vendor of Interest:", 
            font=("Helvetica", 11), bg="#f4f4f4"
        )
        self.select_label.grid(row=0, column=0, padx=5)
        
        
        self.vendor_combo = ttk.Combobox(
            self.selection_frame, values=self.manager.get_vendor_names(), 
            state="readonly", width=25, font=("Helvetica", 10)
        )
        self.vendor_combo.grid(row=0, column=1, padx=5)
        self.vendor_combo.set("Choose a vendor...")
        
    
        self.view_btn = tk.Button(
            root, text="Display Menu", font=("Helvetica", 11, "bold"), 
            bg="#2ca02c", fg="white", command=self.display_menu, relief="flat", padx=10
        )
        self.view_btn.pack(pady=10)
        
        
        self.menu_frame = tk.Frame(root)
        self.menu_frame.pack(pady=15, fill="both", expand=True, padx=20)
        
    
        self.tree = ttk.Treeview(self.menu_frame, columns=("Main Meal", "Price (N)"), show="headings", height=8)
        self.tree.heading("Main Meal", text="Main Meal")
        self.tree.heading("Price (N)", text="Price (₦)")
        self.tree.column("Main Meal", width=250, anchor="w")
        self.tree.column("Price (N)", width=150, anchor="center")
        self.tree.pack(side="left", fill="both", expand=True)
        
    def display_menu(self):
        selected_vendor = self.vendor_combo.get()
        
        if selected_vendor == "Choose a vendor...":
            messagebox.showwarning("Selection Error", "Please select a valid vendor from the dropdown list.")
            return
            
    
        for item in self.tree.get_children():
            self.tree.delete(item)
            
        
        menu = self.manager.get_vendor_menu(selected_vendor)
        
        
        if menu:
            for meal, price in menu.items():
                self.tree.insert("", "end", values=(meal, f"₦ {price}"))


if __name__ == "__main__":
    root = tk.Tk()
    app = VendorApp(root)
    root.mainloop()
```
